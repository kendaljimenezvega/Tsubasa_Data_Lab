```python
import os
import pandas as pd
import soccerdata as sd
from sqlalchemy import create_engine
from dotenv import load_dotenv

load_dotenv()
```

```python
db_url = (
    f"postgresql+psycopg2://{os.getenv('PGUSER')}:{os.getenv('PASSWORD')}"
    f"@{os.getenv('HOST')}:{os.getenv('PORT')}/{os.getenv('DB')}"
)

engine = create_engine(db_url)
```

```python
fb = sd.FBref(leagues="Big 5 European Leagues Combined", seasons=2021)

team_match_stats = fb.read_team_match_stats(stat_type='schedule',opponent_stats=False)


team_match_stats.to_sql(
    "FBref_team_match_stats",
    engine,
    schema="bronze",
    if_exists="replace",
    index=False
)

print(f"Loaded {len(team_match_stats)} rows into bronze.team_season_stats")
```

```python
seasons = list(range(1980, 2026 ))
#print("Fetching seasons:", seasons)
print(type(seasons))
```

```python
for season in seasons:
    print(f"Fetching {season} season data")
    fb = sd.FBref(leagues="Big 5 European Leagues Combined", seasons=season)
    team_match_stats = fb.read_team_match_stats(stat_type='schedule',opponent_stats=False)
    print(season)
```
